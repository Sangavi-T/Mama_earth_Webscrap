```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://mamaearth.in/product-category/combos"
r=requests.get(url)
soup=BeautifulSoup(r.text,"html.parser")
```

```python
product_name=[]
product_link=[]
product_price=[]
product_description=[]
product_review=[]
url_list=[]
```

```python
box = soup.find_all("div", {'class':"Flex-sc-1lsr9yp-0 kWCDdc"}) 
for i in box:
    names = i.find('div', class_="title").text
    price_1=i.find('div',class_="price").text
    desc=i.find("div",class_="subtitle").text
    link_1=i.find("a").get("href")
    reviews=i.find("div",class_="Flex-sc-1lsr9yp-0 cHTzIe rating-review").text
    product_name.append(names)
    product_price.append(price_1)
    product_description.append(desc)
    product_review.append(reviews)
    product_link.append(link_1)
```

```python
domain_name="https://mamaearth.in/"
for i in product_link:
    url=domain_name+i 
    url_list.append(url)
```

```python
dict={"product name":product_name,"Product_url":url_list,"Product_price":product_price,"description":product_description,"review_count":product_review
    }
```

```python
a=pd.DataFrame(dict)
```

```python
a.to_csv("product_mamaearth.csv")
```
